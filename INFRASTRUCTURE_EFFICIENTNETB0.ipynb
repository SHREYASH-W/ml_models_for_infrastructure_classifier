```python
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import numpy as np
import os

# Set data directory
DATA_DIR = "D:/infra_e_net"

# Parameters
IMG_SIZE = (224, 224)
BATCH_SIZE = 16
EPOCHS = 50
LEARNING_RATE = 0.0001

# Enhanced data preprocessing and augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    zoom_range=0.3,
    shear_range=0.3,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest',
    validation_split=0.2
)

# Separate generator for validation (no augmentation)
val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# Create data generators with proper configuration
train_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training',
    shuffle=True,
    seed=42  # Added seed for reproducibility
)

validation_generator = val_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation',
    shuffle=False,
    seed=42  # Added seed for reproducibility
)

print(f"Training samples: {train_generator.samples}")
print(f"Validation samples: {validation_generator.samples}")
print(f"Class indices: {train_generator.class_indices}")

# Calculate class weights to handle imbalanced data
def calculate_class_weights(generator):
    """Calculate class weights for imbalanced dataset"""
    labels = generator.classes
    class_weights = compute_class_weight(
        'balanced',
        classes=np.unique(labels),
        y=labels
    )
    return dict(enumerate(class_weights))

class_weights = calculate_class_weights(train_generator)
print(f"Class weights: {class_weights}")

# Enhanced model architecture
def create_improved_efficientnet_model():
    # Load pre-trained EfficientNet-B0
    base_model = EfficientNetB0(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )
    
    # Freeze base model layers initially
    base_model.trainable = False
    
    # Enhanced classification head
    inputs = base_model.input
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    x = Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.2)(x)
    predictions = Dense(1, activation='sigmoid')(x)
    
    model = Model(inputs=inputs, outputs=predictions)
    
    return model

# Create and compile model
model = create_improved_efficientnet_model()
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss='binary_crossentropy',
    metrics=['accuracy', 'precision', 'recall']
)

print("Model Summary:")
model.summary()

# Enhanced callbacks
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=15,
        restore_best_weights=True,
        verbose=1,
        mode='max'
    ),
    ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.3,
        patience=7,
        min_lr=1e-8,
        verbose=1,
        mode='max'
    ),
    ModelCheckpoint(
        'best_infrastructure_model_v2.keras',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1,
        mode='max'
    )
]

# Phase 1: Training with frozen base model
print("Phase 1: Training with frozen base model...")
try:
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // BATCH_SIZE,
        epochs=25,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // BATCH_SIZE,
        callbacks=callbacks,
        class_weight=class_weights,
        verbose=1
    )
    
    # Phase 2: Fine-tuning with gradual unfreezing
    print("\nPhase 2: Fine-tuning with gradual unfreezing...")
    
    # Unfreeze top layers gradually
    base_model = model.layers[0]
    base_model.trainable = True
    
    # Freeze bottom layers, unfreeze top layers
    for i, layer in enumerate(base_model.layers):
        if i < len(base_model.layers) * 0.8:  # Freeze bottom 80% of layers
            layer.trainable = False
        else:
            layer.trainable = True
    
    # Recompile with lower learning rate
    model.compile(
        optimizer=Adam(learning_rate=LEARNING_RATE/10),
        loss='binary_crossentropy',
        metrics=['accuracy', 'precision', 'recall']
    )
    
    # Reset callbacks with new checkpoint name
    callbacks[2] = ModelCheckpoint(
        'best_infrastructure_model_v2_finetuned.keras',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1,
        mode='max'
    )
    
    # Continue training
    history_fine = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // BATCH_SIZE,
        epochs=25,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // BATCH_SIZE,
        callbacks=callbacks,
        class_weight=class_weights,
        verbose=1
    )
    
    # Combine histories
    history_combined = {
        'loss': history.history['loss'] + history_fine.history['loss'],
        'accuracy': history.history['accuracy'] + history_fine.history['accuracy'],
        'val_loss': history.history['val_loss'] + history_fine.history['val_loss'],
        'val_accuracy': history.history['val_accuracy'] + history_fine.history['val_accuracy']
    }
    
    # Enhanced plotting function
    def plot_training_history(history):
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
        
        # Plot accuracy
        ax1.plot(history['accuracy'], label='Training Accuracy', color='blue')
        ax1.plot(history['val_accuracy'], label='Validation Accuracy', color='red')
        ax1.set_title('Model Accuracy')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Accuracy')
        ax1.legend()
        ax1.grid(True)
        
        # Plot loss
        ax2.plot(history['loss'], label='Training Loss', color='blue')
        ax2.plot(history['val_loss'], label='Validation Loss', color='red')
        ax2.set_title('Model Loss')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Loss')
        ax2.legend()
        ax2.grid(True)
        
        plt.tight_layout()
        plt.show()
    
    plot_training_history(history_combined)
    
    # Comprehensive evaluation
    print("\nComprehensive model evaluation...")
    val_loss, val_accuracy, val_precision, val_recall = model.evaluate(validation_generator)
    
    print(f"Validation Accuracy: {val_accuracy:.4f}")
    print(f"Validation Loss: {val_loss:.4f}")
    print(f"Validation Precision: {val_precision:.4f}")
    print(f"Validation Recall: {val_recall:.4f}")
    
    # Calculate F1 score
    f1 = 2 * (val_precision * val_recall) / (val_precision + val_recall)
    print(f"F1 Score: {f1:.4f}")
    
    # Get predictions for detailed analysis
    validation_generator.reset()
    val_predictions = model.predict(validation_generator, verbose=1)
    val_predictions_binary = (val_predictions > 0.5).astype(int)
    val_true_labels = validation_generator.classes
    
    # Calculate additional metrics
    from sklearn.metrics import classification_report, confusion_matrix
    
    print("\nClassification Report:")
    print(classification_report(val_true_labels, val_predictions_binary, 
                              target_names=['Bad Infrastructure', 'Good Infrastructure']))
    
    print("\nConfusion Matrix:")
    cm = confusion_matrix(val_true_labels, val_predictions_binary)
    print(cm)
    
    # Save the enhanced model
    model.save('infrastructure_efficientnet_b0_enhanced.keras')
    print("Enhanced model saved as 'infrastructure_efficientnet_b0_enhanced.keras'")
    
except Exception as e:
    print(f"Error during training: {e}")
    print("Attempting alternative approach...")
    
    # Alternative: Use tf.data for better control
    def create_tf_dataset(directory, batch_size, img_size, validation_split=0.2):
        """Create tf.data dataset from directory"""
        dataset = tf.keras.utils.image_dataset_from_directory(
            directory,
            validation_split=validation_split,
            subset="training",
            seed=42,
            image_size=img_size,
            batch_size=batch_size,
            label_mode='binary'
        )
        
        val_dataset = tf.keras.utils.image_dataset_from_directory(
            directory,
            validation_split=validation_split,
            subset="validation",
            seed=42,
            image_size=img_size,
            batch_size=batch_size,
            label_mode='binary'
        )
        
        return dataset, val_dataset
    
    try:
        train_ds, val_ds = create_tf_dataset(DATA_DIR, BATCH_SIZE, IMG_SIZE)
        
        # Normalize the data
        normalization_layer = tf.keras.layers.Rescaling(1./255)
        train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
        val_ds = val_ds.map(lambda x, y: (normalization_layer(x), y))
        
        # Configure for performance
        AUTOTUNE = tf.data.AUTOTUNE
        train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
        val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
        
        print("Using tf.data approach for training...")
        
        # Train with tf.data
        history = model.fit(
            train_ds,
            validation_data=val_ds,
            epochs=50,
            callbacks=callbacks,
            verbose=1
        )
        
        # Plot results
        def plot_training_history_simple(history):
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
            
            # Plot accuracy
            ax1.plot(history.history['accuracy'], label='Training Accuracy', color='blue')
            ax1.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
            ax1.set_title('Model Accuracy')
            ax1.set_xlabel('Epoch')
            ax1.set_ylabel('Accuracy')
            ax1.legend()
            ax1.grid(True)
            
            # Plot loss
            ax2.plot(history.history['loss'], label='Training Loss', color='blue')
            ax2.plot(history.history['val_loss'], label='Validation Loss', color='red')
            ax2.set_title('Model Loss')
            ax2.set_xlabel('Epoch')
            ax2.set_ylabel('Loss')
            ax2.legend()
            ax2.grid(True)
            
            plt.tight_layout()
            plt.show()
        
        plot_training_history_simple(history)
        
        # Final evaluation
        results = model.evaluate(val_ds)
        print(f"Final validation results: {results}")
        
        # Save model
        model.save('infrastructure_efficientnet_b0_fixed.keras')
        print("Model saved as 'infrastructure_efficientnet_b0_fixed.keras'")
        
    except Exception as e2:
        print(f"Alternative approach also failed: {e2}")
        print("Please check your data directory structure and ensure it follows the expected format:")
        print("DATA_DIR/")
        print("  ├── class1/")
        print("  │   ├── image1.jpg")
        print("  │   └── image2.jpg")
        print("  └── class2/")
        print("      ├── image3.jpg")
        print("      └── image4.jpg")

# Enhanced prediction function
def predict_infrastructure_enhanced(model, img_path, threshold=0.5):
    """
    Enhanced prediction function with confidence scoring
    """
    try:
        img = tf.keras.preprocessing.image.load_img(img_path, target_size=IMG_SIZE)
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = img_array / 255.0
        
        prediction = model.predict(img_array)[0][0]
        
        if prediction > threshold:
            label = "Good Infrastructure"
            confidence = prediction
        else:
            label = "Bad Infrastructure"
            confidence = 1 - prediction
        
        return label, confidence, prediction
    except Exception as e:
        print(f"Error in prediction: {e}")
        return None, None, None

# Data analysis function
def analyze_dataset_structure(data_dir):
    """Analyze dataset structure and class distribution"""
    try:
        classes = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]
        print(f"\nDataset Structure Analysis:")
        print(f"Data directory: {data_dir}")
        print(f"Found classes: {classes}")
        
        total_images = 0
        for class_name in classes:
            class_path = os.path.join(data_dir, class_name)
            image_files = [f for f in os.listdir(class_path) 
                          if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
            count = len(image_files)
            total_images += count
            print(f"  {class_name}: {count} images")
        
        print(f"Total images: {total_images}")
        
        if total_images == 0:
            print("Warning: No images found! Please check your data directory structure.")
        
    except Exception as e:
        print(f"Error analyzing dataset: {e}")

# Analyze your dataset
analyze_dataset_structure(DATA_DIR)

print("\n=== Model Training Tips ===")
print("1. Ensure your data directory follows the correct structure")
print("2. Check that images are in supported formats (jpg, png, etc.)")
print("3. Verify class names are correct")
print("4. Consider data augmentation for small datasets")
print("5. Monitor training closely for overfitting")
print("6. Try different learning rates if convergence is slow")

print("\n=== Troubleshooting ===")
print("If you encounter the 'Graph execution error':")
print("1. Check TensorFlow version compatibility")
print("2. Ensure data directory structure is correct")
print("3. Try reducing batch size")
print("4. Use tf.data instead of ImageDataGenerator")
print("5. Check for corrupted images in your dataset")
```
